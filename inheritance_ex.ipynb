```python
#%%writefile inhe.py
class papa:
    Fname =" "
    Lname =" "
    Age =" "
    
    def __init__(self,Fname,Lname,Age):
        self.first_name = Fname
        self.last_name = Lname
        self.age = Age
    def intro(self):
        print("Hello My Name is",self.first_name,self.last_name)
        print("\n And my age is", self.age)
class son(papa):

    def __init__(self,Fname,Lname,Age,address):
        papa.__init__(self,Fname,Lname,Age)
        self.add = address
    def myadd(self):
        print("My address is",self.add)

S1 = son('Shoun','Deshpande','30','Pune')
S1.intro()
S1.myadd()

P1 = papa('Sanjay','Deshpande','60')
P1.intro()
```
